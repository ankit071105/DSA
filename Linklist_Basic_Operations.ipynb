```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class Solution:

    # Insert at Beginning
    def insertAtBeginning(self, head, x):
        newNode = Node(x)
        newNode.next = head
        return newNode

    # Insert at End
    def insertAtEnd(self, head, x):
        n = Node(x)
        if head is None:
            return n 
        c = head
        while c.next:
            c = c.next
        c.next = n
        return head

    # Insert at Position (0-based index)
    def insertAtPosition(self, head, pos, x):
        if pos == 0:
            return self.insertAtBeginning(head, x)

        newNode = Node(x)
        curr = head

        for _ in range(pos - 1):
            if curr is None:
                return head
            curr = curr.next

        if curr is None:
            return head

        newNode.next = curr.next
        curr.next = newNode
        return head

    # Delete First Node
    def deleteFirst(self, head):
        if head is None:
            return None
        return head.next

    # Delete Last Node
    def deleteLast(self, head):
        if head is None or head.next is None:
            return None

        curr = head
        while curr.next.next:
            curr = curr.next

        curr.next = None
        return head

    # Delete by Value
    def deleteByValue(self, head, key):
        if head is None:
            return None

        if head.data == key:
            return head.next

        curr = head
        while curr.next and curr.next.data != key:
            curr = curr.next

        if curr.next:
            curr.next = curr.next.next

        return head

    # Delete by Position (0-based)
    def deleteAtPosition(self, head, pos):
        if head is None:
            return None

        if pos == 0:
            return head.next

        curr = head
        for _ in range(pos - 1):
            if curr.next is None:
                return head
            curr = curr.next

        if curr.next:
            curr.next = curr.next.next

        return head

    # Search
    def searchKey(self, head, key):
        curr = head

        while curr:
            if curr.data == key:
                return True
            curr = curr.next

        return False

    # Count Nodes
    def countNodes(self, head):
        count = 0
        curr = head

        while curr:
            count += 1
            curr = curr.next

        return count

    # Reverse Linked List
    def reverse(self, head):
        prev = None
        curr = head

        while curr:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt

        return prev

    # Print Linked List
    def printList(self, head):
        curr = head

        while curr:
            print(curr.data, end=" -> ")
            curr = curr.next

        print("None")
```

```python
# 160. Intersection of Two Linked Lists
class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution(object):
    def getIntersectionNode(self, headA, headB):
        """
        :type head1, head1: ListNode
        :rtype: ListNode
        """
        p1 = headA
        p2 = headB

        while p1 != p2:
            p1 = p1.next if p1 else headB
            p2 = p2.next if p2 else headA

        return p1
```

```python
# 328. Odd Even Linked List
class Solution(object):
    def oddEvenList(self, head):
        if not head or not head.next:
            return head

        odd, even = head, head.next
        e = even

        while even and even.next:
            odd.next = even.next
            odd = odd.next
            even.next = odd.next
            even = even.next

        odd.next = e
        return head
        
```
