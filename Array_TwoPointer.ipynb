```python
# Two Sum with input (2,7,11,15) t=9 
def twosum(n,t):
    i=0
    j=len(n)-1
    while i < j:
        r=n[i]+n[j]
        if r ==t:
            return [i,j]
        if r <t:
            i+=1
        if r>t:
            j-=1
        if i==j:
            break
    return "ArithmeticError: No two sum solution found"
twosum([2,7,11,15],26)

            
```

```python
# Three Sum with input (2,7,11,15) t=26


# Brute force solution
def threesum(n,t):
    n.sort()
    s=len(n)
    b=[]
    for i in range(s-2):
        for j in range(i+1,s-1):
            for k in range(j+1,s):
                if n[i]+n[j]+n[k]==t:
                    b.append([n[i],n[j],n[k]])
    if len(b)==0:
        return "ArithmeticError: No three sum solution found"
    return b

threesum([2,7,11,15],12)

```
