```python
# 876. Middle of the Linked List

class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution(object):
    def middleNode(self, head):
        f=s=head
        while f is not None and f.next is not None:
            s=s.next
            f=f.next.next
        return s
```

```python
# 141. Linked List Cycle Brute Force 

def hasCycle(self, head):
        c = head
        s = set()
        while c:
            if c in s:
                return True
            s.add(c)
            c = c.next
        return False
```

```python
# 141. Linked List Cycle Optimal Code

class Solution(object):
    def hasCycle(self, head):
        f=s=head
        while f is not None and f.next is not None:
            s=s.next
            f=f.next.next
            if f==s:
                return True
        return False
```

```python
# 142. Linked List Cycle II Node value return

class Solution(object):
    def detectCycle(self, head):
        f = s = head
        while f is not None and f.next is not None:
            s = s.next
            f = f.next.next
            if s == f:
                break
        else:
            return None

        s = head
        while s != f:
            s = s.next
            f = f.next

        return s
```

```python
# 19. Remove Nth Node From End of List

class Solution(object):

    def removeNthFromEnd(self, head, n):
        d = ListNode(0)
        d.next = head
        fast = slow = d
        for _ in range(n + 1):
            fast = fast.next
        while fast:
            fast = fast.next
            slow = slow.next
        slow.next = slow.next.next
        return d.next
```
