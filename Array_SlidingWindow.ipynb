```python
# Max Sum Sub Array with Brute Force Approach
def max_sub(arr,k):
    n=len(arr)
    max_sum=0
    for i in range(n-k+1):
        sum=0
        for j in range(i,i+k):
            sum+=arr[j]
        max_sum=max(max_sum,sum)
    return max_sum

max_sub([1,2,3,4,5],2)
    
```

```python
# Max Sum Sub Array with Sliding Window Approach
def max_sub(arr,k):
    n=len(arr)
    max_sum=0
    sum=0
    for i in range(k):
        sum+=arr[i]
    max_sum=sum
    for j in range(k,n):
        sum+=arr[j]-arr[j-k]
        max_sum=max(max_sum,sum) 
    return max_sum

max_sub([1,2,3,4,5],2)
```

```python
# Maximum Consecutive Ones with Sliding Window Approach
def max_consecutive_ones(arr):
    s=c=0
    n=len(arr)
    for i in range(n):
        if arr[i]==1:
            c+=1
        else:
            s=max(s,c)
            c=0
    return max(s,c)
max_consecutive_ones([1,1,0,1,1,1])
```

```python
# maximum consecutive ones with at most k zeros with Brute Force approach
def max_consecutive_ones(arr,k):
    n=len(arr)
    max_ans=0
    for i in range(n):
        s=0
        for j in range(i,n):
            if arr[j]==0:
                s+=1
            if s>k:
                break
            max_ans=max(max_ans,j-i+1)
    return max_ans
max_consecutive_ones([1,1,0,0,0,1,0,1,0,1,0,1,1],3)
```

```python
# maximum consecutive ones with at most k zeros with Sliding Window approach
def max_consecutive_ones(a,k):
    n=len(a)
    l=m=z=0
    for i in range(n):
        if a[i]==0:
            z+=1
        if z>k:
            if a[l] ==0:
                z-=1
            l+=1
        m=max(m,i-l+1)
    return m
max_consecutive_ones([1,1,0,0,0,1,0,1,0,1,0,1,1],3)
    
```

```python
# Subarray Product Less Than K  Sliding Window Approach
def sub_array(a,k):
    n=len(a)
    p=1
    c=0
    l=0
    for i in range(n):
        p *=a[i]
        while p >=k:
            p /= a[l]
            l+=1
        c+=i-l+1
    return c
sub_array([10,5,2,6] , 100)

    
```

```python
# 209 Minimum Size Subarray Sum
def min_sum(a,t):
    ans=float('inf')
    s=l=0
    for r in range(len(a)):
        s+=a[r]
        while s >= t:
            ans=min(ans,r-l+1)
            s-=a[l]
            l+=1
    return 0 if ans == float('inf') else ans
a=[2,3,1,2,4,3]
t=7
min_sum(a,t)
```

```python
#  Subarrays with K Different Integers with optimized code 
def atMost(nums, k):
    l = 0
    ans = 0
    freq = {}

    for r in range(len(nums)):
        freq[nums[r]] = freq.get(nums[r], 0) + 1

        while len(freq) > k:
            freq[nums[l]] -= 1
            if freq[nums[l]] == 0:
                del freq[nums[l]]
            l += 1

        ans += r - l + 1

    return ans

def subarraysWithKDistinct(nums, k):
    return atMost(nums, k) - atMost(nums, k - 1)

nums = [1,2,1,2,3]
print(subarraysWithKDistinct(nums, 2))
```

```python
# 239. Sliding Window Maximum  With Brute Force
def max_slide(a,k):
    n=len(a)
    r=[]
    for i in range(n-k+1):
        m=a[i]
        for j in range(i,i+k):
            m=max(m,a[j])
        r.append(m)
    return r
a = [1,3,-1,-3,5,3,6,7]
k = 3
max_slide(a,k)
        
```

```python
# 239. Sliding Window Maximum  With Optimal code 
from collections import deque
def max_slide(a,k):
    n=len(a)
    r=[]
    q=deque() 
    for i in range(n):
        while q and q[0] <= i - k:
            q.popleft()
        while q and a[q[-1]] <= a[i]: 
            q.pop()
        q.append(i)
        if i >= k-1:
            r.append(a[q[0]])
        
    return r
a = [1,3,-1,-3,5,3,6,7]
k = 3
max_slide(a,k)
```
