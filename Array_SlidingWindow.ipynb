```python
# Max Sum Sub Array with Brute Force Approach
def max_sub(arr,k):
    n=len(arr)
    max_sum=0
    for i in range(n-k+1):
        sum=0
        for j in range(i,i+k):
            sum+=arr[j]
        max_sum=max(max_sum,sum)
    return max_sum

max_sub([1,2,3,4,5],2)
    
```

```python
# Max Sum Sub Array with Sliding Window Approach
def max_sub(arr,k):
    n=len(arr)
    max_sum=0
    sum=0
    for i in range(k):
        sum+=arr[i]
    max_sum=sum
    for j in range(k,n):
        sum+=arr[j]-arr[j-k]
        max_sum=max(max_sum,sum) 
    return max_sum

max_sub([1,2,3,4,5],2)
```

```python
# Maximum Consecutive Ones with Brute Force Approach
def max_consecutive_ones(arr):
    n=len(arr)
    max_ans=0
    for i in range(n):
        for j in range(i,n):
            if arr[j]==0:
                s=j-i
                max_ans=max(max_ans,s)
                break
    return max_ans
max_consecutive_ones([1,1,0,1,1,1,0,1,1])
            
        
    
```

```python
# Maximum Consecutive Ones with Sliding Window Approach
def max_consecutive_ones(arr):
    s=c=0
    n=len(arr)
    for i in range(n):
        if arr[i]==1:
            c+=1
        else:
            s=max(s,c)
            c=0
    return max(s,c)
max_consecutive_ones([1,1,0,1,1,1])
```
