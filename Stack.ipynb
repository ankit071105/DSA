```python
 # 496. Next Greater Element I Brute Force Approach
def next_great(a):
    n=len(a) 
    ans=[-1]*n
    for i in range(n-1):
        for j in range(i+1, n):
            if a[j]>a[i]:
                ans[i]=a[j]
                break
    return ans
a=[1, 3, 2, 4]
next_great(a)
                
 
```

```python
 # 496. Next Greater Element I Optimal Force Approach
def next_great(a):
    n=len(a) 
    ans=[-1]*n
    s=[]
    for i in range(n-1,-1,-1): # iterate from right to left
        while s and s[-1] <= a[i]:
            s.pop()
        if s:
            ans[i]=s[-1]
        s.append(a[i])
    return ans
a=[1, 3, 2, 4]
next_great(a)
```
